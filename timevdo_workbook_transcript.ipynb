```python
def clean_text(text):
    temp = re.sub(CLEANR, '', text)
    temp = re.sub(CLEANR2, '', temp)
    temp = temp.replace('\n', ' ')
    temp = re.sub(CLEANR3, ' ', temp).strip()
    return temp
```

```python
def get_bill_data(bill):
    if bill not in set(rolldf['bill_number']):
        return None
    return rolldf[rolldf['bill_number'] == bill]
```

```python
# Load the glove mode
vec_txt_file_lines = open('glove.6B.100d.txt', 'r').readlines()

glove = {}

for line in vec_txt_file_lines:
    line_split = line.split(' ')
    key = line_split[0]
    vec = np.array([float(v) for v in line_split[1:]])
    glove[key] = vec
```

```python
with open("./congress/mentions_of_bills.json") as f:
	data = json.load(f)

bills = set()
bill_keys = []
data_arr = []

BILL_REGEX = re.compile(r"(H\.R\. \d+|H\. Res\. \d+|S\. \d+|S\. Res\. \d+)")
BILL_KEY = re.compile(r"[\s\.]")
for item in data:
	b = (re.search(BILL_REGEX, item['transcript']).group(1))
	bkey = re.sub(BILL_KEY, "", b)
	bdata = get_bill_data(bkey)
	if bdata is not None:
		transcript_vecs = []
		clean_transcript = clean_text(item['transcript'])
		for w in clean_transcript.split(' '):
			if w.lower() in glove:
				transcript_vecs.append(glove[w.lower()])

		data_arr.append([bkey, transcript_vecs, int(max(bdata['yea_count']))])
```
